```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(42)
```

```python
np.random.seed(42)
xs = np.random.uniform(-1, 1, 20)
ys = np.random.uniform(-1, 1, 20)
point_1 = (-0.75, 0.75)
point_2 = (0.75, -0.75)
```

```python
fig, ax = plt.subplots()
sampled_x, sampled_y = xs[:5], ys[:5]
ax.scatter(xs, ys, label="Examples")
ax.legend()
ax.set_title("Projection of our available examples")
```

```python
fig, ax = plt.subplots()
sampled_x, sampled_y = xs[:5], ys[:5]
ax.scatter(xs[5:], ys[5:], label="Examples")
ax.scatter(sampled_x, sampled_y, label="Randomly selected examples", color="red")
ax.legend()
ax.set_title("Selecting random examples")
```

```python
fig, ax = plt.subplots()
ax.scatter(xs, ys, label="Examples")
ax.plot(*point_1, 'ro')
# draw an arrow to the point_1 and label it as "example 1"
ax.annotate('Item 1', xy=point_1, xytext=(0, 0.75),
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.plot(*point_2, 'ro')
# draw an arrow to the point_2 and label it as "example 2"
ax.annotate('Item 2', xy=point_2, xytext=(-0.25, -0.60),
            arrowprops=dict(facecolor='black', shrink=0.05))
# draw a circle around the point_1 
circle = ax.add_artist(plt.Circle(point_1, 0.3, fill=False))
# draw a circle around the point_2
circle = ax.add_artist(plt.Circle(point_2, 0.3, fill=False))
ax.set_xlim(-1.25, 1.25)
ax.set_ylim(-1.25, 1.25)
ax.legend()
ax.set_title("Selecting examples with nearest neighbors")
```
